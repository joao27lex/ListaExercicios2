```python
#QUESTÃO 3
class Retangulo:
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def mudaLado(self, novaBase, novaAltura):
        self.base = novaBase
        self.altura = novaAltura

    def retornaLado(self):
        print(f'A base é {self.base} e a altura é {self.altura}')

    def calcArea(self):
        return self.base * self.altura

    def calcPerimetro(self):
        return 2 * self.base + 2 * self.altura

b = int(input('Digite o tamanho do comprimento do local: '))
h = int(input('Digite o tamanho da largura do local: '))
retan = Retangulo(b, h)

area = retan.calcArea()
perimetro = retan.calcPerimetro()

piso = float(input('Digite o tamanho do piso (m²): '))
rodape = float(input('Digite o tamanho do rodapé (m): '))

qtdPiso = area / piso
qtdRodape = perimetro / rodape

print(f'Será possível utilizar {qtdPiso} pisos e {qtdRodape} rodapés')
```

```python
#QUESTÃO 6
class TV:
    def __init__(self, canal, volume):
        self.canal = canal
        self.volume = volume

    def trocaCanal(self, novoCanal):
        if 1 <= novoCanal <= 20:
            self.canal = novoCanal

    def aumentaVolume(self):
        if self.volume<100:
            self.volume += 1
        else:
            print('Você já está no volume máximo!')

    def diminuiVolume(self):
        if self.volume<0:
            self.volume -= 1
        else:
            print('Não dá para abaixar o volume mais que isso, já está mudo!!!')

    def mostra(self):
        print(f'O canal que está sendo transmitido é o {self.canal}')
        print(f'O volume atual é {self.volume}')

tv = TV(20, 99)
tv.mostra()

print('=-' * 20)

tv.trocaCanal(13)
tv.aumentaVolume()
tv.mostra()
```

```python
class Ponto:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def mostra(self):
        print(f'{self.x}, {self.y}')

    def move(self, dx, dy):
        self.x += dx
        self.y += dy

class Retangulo:
    def __init__(self, largura, altura):
        self.largura = largura
        self.altura = altura
        self.ponto_inicial = Ponto(0, 0)

    def centro(self):
        x_centro = self.largura / 2
        y_centro = self.altura / 2
        ponto_centro = Ponto(x_centro, y_centro)
        ponto_centro.mostra()

    def move(self, dx, dy):
        self.ponto_inicial.move(dx, dy)
        print(f'({self.ponto_inicial.x}, {self.ponto_inicial.y})')

retangulo = Retangulo(8, 5)
retangulo.move(3, 8)
retangulo.move(5, 2)
```
